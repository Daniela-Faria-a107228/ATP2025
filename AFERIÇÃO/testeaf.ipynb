Especifique as seguintes funções recursivas. Caso não consiga, pode apresentar uma solução iterativa
que será penalizada em 50%:

a) contaEspacos: defina uma função que recebe uma string e devolve o número de espaços na string.

b) ordenada?: defina uma função que recebe uma lista como argumento e devolve o valor booleano
verdadeiro se a lista estiver ordenada e falso caso contrário.

c) existeElem: defina uma função que recebe uma lista e um elemento e devolve True ou False caso o
elemento exista ou não na lista.


```python
#a) 
#RECURSIVA
def contaEspacos(s):
    if s == '':
        return 0
    if s[0] == ' ':
        return 1 + contaEspacos(s[1:])
    else:
        return contaEspacos(s[1:])

contaEspacos('! !')

#ITERATIVA
def contaEspacos1(s):
    soma = 0
    for espaco in s:
        if espaco == ' ':
            soma = soma + 1
    return soma

contaEspacos1('! !')
```

```python
#b)
#RECURSIVA
def ordenada(lista):
    if len(lista) <= 1:
        return True
    if lista[0] > lista[1]:
        return False
    return ordenada(lista[1:])
    
ordenada([1,2,3,4])
```

```python
#c)
#RECURSIVA
def existeElem(lista,elem):
    if lista == []:
        return False
    if lista[0] == elem:
        return True
    else:
        return existeElem(lista[1:], elem)
    
existeElem([1,2],2)

#ITERATIVA
def existeElem1(lista,elem):
    encontrado = False
    for carater in lista:
        if carater == elem:
            encontrado = True
    return encontrado

    
```

a) maiorDivisor: defina uma função que recebe um número N, e calcula o maior número de um digito (1 a
9) que divide cada número de entre 1 e N. O resultado será um dicionário.

```python
def maiorDivisor(N):
    res = {}
    for num in range(1,N+1):
        maior = 1
        for divisor in range(1,10):
            if num % divisor == 0:
                maior = divisor
        res[num] = maior
    return res

maiorDivisor(20)
```

b) mmc: defina uma função que calcula o mínimo múltiplo comum entre dois números inteiros.


```python
def mmc(a, b):
    i = 1
    resultado = 0
    while resultado == 0:   # vamos parar quando encontrar o MMC
        if i % a == 0 and i % b == 0:
            resultado = i
        i = i + 1
    return resultado

mmc(12,16)
```

c) f2c: Converta um registo de temperaturas Fahrenheit, num novo registo em que a cada temperatura
Fahrenheit associas uma temperatura em Celsius.

```python
def f2c(fahrenheit):
    celsius = {}
    for chave in fahrenheit:
        f = fahrenheit[chave]
        c = (f - 32) / 1.8
        celsius[chave] = (f, c)
    return celsius

f2c({'t1': -30, 't2': -20, 't3': -10, 't4': 0})
```

d) reduxInt: defina uma função que recebe um número inteiro positivo como argumento e vai adicionar
repetidamente os seu dígitos até obter apenas um dígito que é retornado como resultado.

```python
def reduxInt(n):
    while n >= 10:  
        soma = 0
        for digito in str(n):   
            soma = soma + int(digito) 
        n = soma
    return n

reduxInt(998)
```

d) existe_medico: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o nome de um médico, e verifica se o médico existe nessa unidade hospitalar, devolvendo True caso exista e False em caso contrário.

In: print(existe_medico(hospital, "Unidade de Cardiologia", "Dra.Maria"))                                 Out: True

```python
def existe_medico(Hospital,Unidade,Medico):
    cond = False
    for unidade in Hospital:
        if unidade[0] == Unidade:
            for medico in unidade[1]:
                if medico[1] == Medico:
                    cond = True
    return cond

existe_medico(hospital, 'Unidade de Cardiologia', 'Dra. Maria')
```

e) adicionar_medico: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o nome de um médico, e caso este não exista na unidade, acrescenta-o e devolve o novo hospital com o médico inserido.

In: print(adicionar_medico(hospital, "Unidade de Cardiologia", "Dra.Paula")) Out: [("Unidade de Cardiologia",[(1, "Dr. João"), (2, "Dra. Maria"), (3,"Dra. Paula")] ...)]

```python
def adicionar_medico(Hospital,Unidade,Medico):
    for unidade in Hospital:
        if unidade[0] == Unidade:
            if Medico not in unidade[1]:
                unidade[1].append(Medico)
    return Hospital

adicionar_medico(hospital, "Unidade de Cardiologia", "Dra.Paula")
```

f) ocupar_cama: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o identificador de uma cama, e caso a cama esteja livre devolve um novo hospital com a cama ocupada (não faz nada se a cama estiver ocupada).

In: print(ocupar_cama(hospital, "Unidade de Cardiologia", 101))
Out: [("Unidade de Cardiologia",[...], [(101, True), (102, True)])...]

```python
def ocupar_cama(Hospital, Unidade, IDcama):
    for unidade in Hospital:
        if unidade[0] == Unidade:
            camas = unidade[2]
            for i in range(len(camas)):
                numero, ocupada = camas[i]
                if numero == IDcama and not ocupada:
                    camas[i] = (numero, True)
    return Hospital
```

i) camas_ocupadas: Defina uma função que recebe um hospital e devolve a percentagem de camas ocupadas do total de camas no hospital.

```python
def camas_ocupadas(Hospital):
    total = 0
    ocupadas = 0
    for unidade in Hospital:
        camas = unidade[2]
        for i in range(len(camas)):
            total = total + 1
            if camas[i][1] == True:
                ocupadas = ocupadas + 1
            if total == 0:
                return 0
    return (ocupadas / total) * 100

camas_ocupadas(hospital)
```

j) unidades_ocupacao_critica: Defina uma função que recebe um hospital e devolve uma lista de tuplos (pares), em que o primeiro elemento do tuplo é o nome de uma unidade hospitalar e o segundo é a percentagem de camas ocupadas nessa unidade, mas apenas deverão constar tuplos em que a percentagem é maior ou igual a 80%.

```python
def unidades_ocupacao_critica(Hospital):
    res = []
    for i in range(len(Hospital)):
        unidade = Hospital[i]
        nome = unidade[0]
        camas = unidade[2]
        total = len(camas)
        if total > 0:  
            ocupadas = 0
            for j in range(len(camas)):
                if camas[j][1] == True:
                    ocupadas = ocupadas + 1
            percentagem = (ocupadas / total) * 100
            if percentagem >= 80:
                res.append((nome, percentagem))
    return res

unidades_ocupacao_critica(hospital)
```
